# Równanie struny

Funkcja dwóch zmiennych $u(x, t)$ - równanie różniczkowe cząstkowe (PDE) ze zmienną **t**
  
$$ \frac{\partial^{2}u}{\partial t^{2}} - c^2 \frac{\partial^{2}u}{\partial x^{2}} = 0 \tag{1}$$  

Równoważny zapis: $u_{tt} - c^2 u_{xx} = 0$

Dodatkowe założenia opisujące fizyczne zagadnienie równania struny:  
- długość struny L, zmienna t opisująca czas $ x \in [0, L], t \geq 0$
- $c^2$ - stała c jest związana z napięciem struny, $c > 0$
- warunki brzegowe: $u(0, t) = u(L, t) = 0$
- warunki początkowe: 
    - $u(x, 0) = f(x)$ (początkowe położenie struny jest opisywane przez pewną funkcję f(x))
    - $u_t(x, 0) = 0$ (początkowa prędkość struny)

[Wizualizacja zagadnienia](https://thumbs.gfycat.com/MammothLongBigmouthbass-max-1mb.gif)

**Metoda rozdzielania zmiennych**: Zakładamy, że szukana funkcja daje się wyrazić jako iloczyn:  

$$ u(x, y) = X(x)T(t) \tag{2}$$  


Podstawiając $(2)$ do $(1)$, otrzymujemy:  

$$ X(x)T''(t) + c^2 X(x)''T(t) = 0 $$
  
po przekształceniu i wprowadzeniu stałej separacji $\lambda$ :  
  
$$ \frac{X''(x)}{X(x)} = \frac{T''(t)}{-c^2 T(t)} = -\lambda $$

Uzyskaliśmy w ten sposób dwa równania różniczkowe zwyczajne:

$$ X''(x) + \lambda X(x) = 0 \tag{3}$$  
$$ T''(t) + c^2 \lambda T(t) = 0 \tag{4}$$

Możemy teraz rozwiązać obydwa te równania za pomocą równań charakterystycznych.

```python
x, t, lmbd, r = symbols('x t lambda r')
u, X, T = map(Function, 'uXT')
```

### Pierwszy przypadek: $\lambda > 0 $  

Równanie charakterystyczne dla $(3): r^2 + \lambda = 0, r_1 = i\sqrt{\lambda}, r_2 = -i\sqrt{\lambda}$

Rozwiązanie:  
$X(x) = C_1 \cos(\sqrt{\lambda} x) + C_2 \sin(\sqrt{\lambda} x) $

Zanim zaczniemy szukać rozwiązania $T(t)$, wyznaczymy $\lambda$ na podstawie $X(x)$.

### Pozostałe przypadki ($\lambda=0, \lambda<0 $) prowadzą do tzw. trywialnych rozwiązań, gdzie wszystkie stałe wynoszą 0.

_______________________________________________________
Mamy rozwiązanie ogólne na $X(x)$ i warunki brzegowe, które ma spełniać ta funkcja: $X(0) = 0, X(L) = 0$. Uwzględniając te warunki, otrzymujemy:

**Pierwszy warunek:**  
$X(0) = C_1 cos(0) + C_2 sin(0) = C_1 * 1 + C_2 * 0 = C_1 + 0 = 0 => C_1 = 0$  
$X(x) = C_2 sin(\sqrt{\lambda}x)$

**Drugi warunek:**  
$X(L) = C_2 sin(\sqrt{\lambda}L) = 0$

Funkcja $sin(\theta)$ ma wartośc 0 dla argumentów $\theta$, które są całkowitymi wielokrotnościami liczby $\pi$, stąd:  
$\sqrt{\lambda}L = n\pi$  
.  
$\sqrt{\lambda} = \frac{n\pi}{L}, n = 1, 2, 3, ... $

Ostateczne rozwiązanie dla $X(x)$ to **zbiór rozwiązań** zależnych od n:  
$X_n(x) = C_2 sin(\frac{n\pi x}{L})$

_____________________________________________
Rozwiązanie dla $T(t)$ (uwaga na oznaczenia stałych - to inne wartości niż stałe powyżej):  
$T(t) = C_3 \cos(c \sqrt{\lambda} t) + C_4 \sin(c \sqrt{\lambda} t) $

```python
# rozwiązanie dla T(t) z pomoca Sympy
c, lmbd = symbols('c lambda', positive = True)

dsolve(Eq(diff(T(t), t, 2) + c**2*lmbd*T(t), 0))
```


Podstawiając znalezione wcześniej $\lambda$ (możemy je ponownie wykorzystać, bo to ta sama wartość, zgodnie z założeniem stałej separacji):

$T_n(t) = C_3 \cos(c \frac{n\pi x}{L} t) + C_4 \sin(c \frac{n\pi x}{L} t) $

## Znaleziona funkcja $u(x, t)$:

$$u(x, t) = X(x)T(t) = C_2 sin(\frac{n\pi x}{L}) (C_3 \cos(c \frac{n\pi x}{L} t) + C_4 \sin(c \frac{n\pi x}{L} t))$$

Możemy zredukować liczbę wystepujących stałych, pamiętając, że stała wartość pomnożona przez inną stałą daje nową stałą:

$$u(x, t) = X(x)T(t) = sin(\frac{n\pi x}{L}) (C_1 \cos(c \frac{n\pi x}{L} t) + C_2 \sin(c \frac{n\pi x}{L} t))$$

Dalsza redukcja stałych: uwzględnienie warunku $u_t(x, 0) = 0$

Najpierw znajdujemy $u_t$:
  
$$ u_t(x, t) = X(x)T'(t) = sin(\frac{n\pi x}{L}) (-C_1 \frac{n\pi c}{L} \sin(\frac{n\pi c}{L} t) + C_2 \frac{n\pi c}{L} \cos(\frac{n\pi c }{L} t) $$


następnie do znalezionej pochodnej podstawiamy $t = 0$.
  
$ u_t(x, 0) = sin(\frac{n\pi x}{L}) (-C_1 \frac{n\pi c}{L} \sin(\frac{n\pi c}{L} 0) + C_2 \frac{n\pi c}{L} \cos(\frac{n\pi c }{L} 0) = 0 $  
_  
$ \sin(\frac{n\pi c}{L} 0) = \sin(0) = 0 $  
$ \cos(\frac{n\pi c }{L} 0) = \cos(0) = 1 $  
_  
$ u_t(x, 0) = sin(\frac{n\pi x}{L}) (0 + C_2 \frac{n\pi c}{L}) = 0 $ // wniosek: $C_2 = 0$  
## Ostatecznie:  
$u(x, t) = C_1 \cos \frac{n\pi c t}{L} \sin \frac{n\pi x}{L}$

# Zasada superpozycji

Zasada superpozycji: w przypadku r. r. liniowego, suma rozwiązań jest też rozwiązaniem równania.  
Możemy zapisać:  
$$u(x, t) = \sum_{n=1}^{\infty} C_n \cos\frac{n\pi c t}{L} \sin\frac{n\pi x}{L}$$  
Wracamy do warunku początkowego $u(x, 0) = f(x)$:  
$$u(x, 0) = f(x) = \sum_{n=1}^{\infty} C_n \sin\frac{n\pi x}{L}$$  
  
Możemy znaleźć wartości stałych $C_n$, znajdując współczynniki Fouriera funkcji $f(x)$.  
Szereg Fouriera:  
$$f(x) = \frac{a_0}{2} + \displaystyle\sum_{k=1}^{\infty}
\left[ a_k \cos\left(\frac{2\pi k x}{T}\right) + b_k \sin\left(\frac{2\pi k x}{T}\right) \right]$$  
Dla funkcji nieparzystej (tzn. $f(x) = -f(-x)$; przyjmujemy takie rozszerzenie funkcji, żeby ten warunek był spełniony) współczynniki $a_k$ są zerowe:  
$$f(x) = \sum_{k=1}^{\infty} b_k \sin(\frac{2\pi k x}{T})$$   
Wzór na obliczenie współczynników $b_k$ dla przedziału $T = 2\pi$:  
$$ b_k = \frac{1}{\pi}\int_{-\pi}^{\pi} f(x)\sin kx\; dx, k=1,2,3,\ldots$$  
Ostatecznie $C_n$ znajdujemy z:  
$$ C_n = \frac{2}{L}\int_{0}^{L} f(x)\sin \frac{n\pi x}{L} dx, k=1,2,3,\ldots$$
