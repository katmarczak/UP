## Definicja szeregu Fouriera

$f(x) = a_0 + \displaystyle\sum_{k=1}^{\infty}
\left[ a_k \cos\left(\frac{2\pi k x}{T}\right) + b_k \sin\left(\frac{2\pi k x}{T}\right) \right]$,

$\displaystyle t_0<x<t_0+T$

$\displaystyle a_0 = \frac{1}{T}\int_{t_0}^{t_0+T} f(x)dx$ - składowa stała

$ a_k = \displaystyle\frac{2}{T}\int_{0}^{T} f(x)\cos\left(\frac{2\pi k x}{T}\right)dx$, $k=1,2,3,\ldots$

$ b_k = \displaystyle\frac{2}{T}\int_{0}^{T} f(x)\sin\left(\frac{2\pi k x}{T}\right)dx$, $k=1,2,3,\ldots$

$f(x) = \frac{a_0}{2} + \displaystyle\sum_{k=1}^{\infty}
\left[ a_k \cos\left(\frac{2\pi k x}{T}\right) + b_k \sin\left(\frac{2\pi k x}{T}\right) \right]$,

$\displaystyle t_0<x<t_0+T$

$ a_k = \displaystyle\frac{2}{T}\int_{0}^{T} f(x)\cos\left(\frac{2\pi k x}{T}\right)dx$, $k=0,1,2,3,\ldots$

$ b_k = \displaystyle\frac{2}{T}\int_{0}^{T} f(x)\sin\left(\frac{2\pi k x}{T}\right)dx$, $k=1,2,3,\ldots$

**Przypadek $T=2\pi$**

$f(x) = \frac{a_0}{2} + \displaystyle\sum_{k=1}^{\infty}
\left( a_k \cos kx + b_k \sin kx \right)$, $-\pi<x<\pi$

$ a_k = \displaystyle\frac{1}{\pi}\int_{-\pi}^{\pi} f(x)\cos kx\; dx$, $k=0,1,2,3,\ldots$

$ b_k = \displaystyle\frac{1}{\pi}\int_{-\pi}^{\pi} f(x)\sin kx\; dx$, $k=1,2,3,\ldots$

* Jeżeli funkcja $f$ jest parzysta w przedziale $(-\pi,\pi)$  $\left[f(x)=f(-x)\right]$, to $b_k=0$
* Jeżeli funkcja $f$ jest nieparzysta w przedziale $(-\pi,\pi)$ $\left[f(-x)=-f(x)\right]$, to $a_k=0$

## Implementacja z wykorzystaniem całkowania numerycznego

### Całkowanie numeryczne

```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt
```

```python
quad(np.sin, 0, 2*np.pi)
```

```python
quad(np.sin, 0, np.pi)
```

```python
def f(x):
    return x
```

```python
quad(f, 0, 10)
```

```python
def fourier_series_factory(f, n, T=2*np.pi, t0=-np.pi):
    
    ak = 2/T * np.array([
            quad(lambda x: f(x)*np.cos(2*np.pi*k*x/T), t0, t0+T)[0]
        for k in range(n+1) ])
    bk = 2/T * np.array([
            quad(lambda x: f(x)*np.sin(2*np.pi*k*x/T), t0, t0+T)[0]
        for k in range(n+1) ])
    ak[0] = ak[0]/2
    k = np.arange(n+1)
    
    def fourier_series(x, m=n):
        y = ak*np.cos(2*np.pi*k*x/T) + bk*np.sin(2*np.pi*k*x/T)
        return sum(y[:m+1])
    
    return fourier_series
```

```python
n = 100
fs = fourier_series_factory(f, n)
fs = np.vectorize(fs)
```

```python
# vectorize - dekorator, dzięki któremu nasza własna funkcja zachowuje się jak funkcje z biblioteki numpy,
# tzn. może przyjmować i zwracać tablice
def bigger(a, b):
    if a >= b:
       return a
    else:
       return b

# bigger([1, 2, 3, 4], 3)

bigger = np.vectorize(bigger)
bigger([1,2,3,4], 3)
```

```python
x = np.linspace(-np.pi, np.pi, 500)

fig, ax = plt.subplots(figsize=(7,7))

ax.plot(x, fs(x, n))
ax.grid(True)
ax.set_aspect('equal')
plt.show()
```

```python
fs_approx = np.array([fs(x, k) for k in range(0, n)])
```

```python
n_sums = 1

fig, ax = plt.subplots(figsize=(7,7))

ax.plot(x, fs_approx[n_sums])
ax.grid(True)
ax.set_aspect('equal')
plt.show()
```
