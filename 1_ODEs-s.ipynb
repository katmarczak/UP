# Modelowanie stygnięcia układu

### Zagadnienie

Prawo stygnięcia (Newton). Szybkość, z jaką układ stygnie jest proporcjonalna do różnicy temperatur pomiędzy układem a otoczeniem.

Zapis prawa w postaci równania różniczkowego zwyczajnego:

$T'(t) = -k(T(t) - T_a)$

gdzie
* $T(t)$ - temperatura układu w czasie t;
* $T_0$ - początkowa temperatura układu (warunek początkowy równania);
* $T_a$ - temp. otoczenia;
* $t$ - czas;
* $k$ - stała dla danego układu.
