Stan początkowy powinien odzwierciedlać szarpnięcie struny, w którym została ona podniesiona w punkcie $p=0.124$ na wysokość $h=0.056$.

Wskazówka. Zdefiniuj funkcję jako funkcję jednej zmiennej $f(x)$ zależną od globalnych zmiennych $L, p, h$. Funkcja będzie składała się z dwóch prostych przechodzących odpowiednio przez punkty $(0,0)$, $(p, h)$ oraz $(p, h)$, $(L, 0)$.

```python
L = 1
p = 0.124
h = 0.056

@np.vectorize
def f(x):
    if x < p*L:
        return h/p * x
    else:
        return h/(p-L)*x - h*L/(p-L)
```
