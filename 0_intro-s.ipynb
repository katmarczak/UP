### 2.3. Równania różniczkowe

Aby rozwiązać równanie różniczkowe w Sympy, wprowadzamy funkcję jako kolejną zmienną symboliczną.

```python
#f = symbols('f', cls=Function) # zaznaczamy, że ta zmienna nalezy do klasy funkcji
f = Function('f')
```

```python
f(x).diff(x) # symboliczne wyrażenie pochodnej od funkcji
f(x).diff(x, 2) # pochodna drugiego stopnia
```

```python
integrate(f(x).diff(x))
```

Równania tworzymy za pomocą *Eq()*, jako argumenty przekazując *lewą stronę równania (lhs)* oraz *prawą stronę (rhs)*.

```python
Eq(x+1, 0) # prosty przykład równania x+1=0
solve(Eq(x+1, 0))
```

Przykład: rozwiązanie równania różniczkowego

$$\frac{d^{2}}{d x^{2}} f{\left(x \right) - 2\frac{d}{d x} f{\left(x \right)} + f(x)} = \sin{\left(x \right)} $$

```python
diff_eq = Eq(f(x).diff(x, 2) - 2*f(x).diff(x) + f(x), sin(x))
```

```python
# uwaga na różnicę między funkcjami solve() i dsolve()
sol = dsolve(diff_eq, f(x))
sol
```

Jakiego typu rozwiązanie mamy powyżej? Czym jest $C_1, C_2$ w rozwiązaniu?

Możemy sprawdzić poprawność rozwiązania przez podstawienie znalezionej funkcji do równania powyżej.  
Uwaga: uzyskane rozwiązanie ma postać wyrażenia _lhs_ = _rhs_ , chcąc policzyć pochodną funkcji, musimy odwołać się do zawartości prawej strony (rhs)

```python
sol.rhs
```

**Przykład**: Pewna populacja zwierząt z upływem czasu rośnie w tempie proporcjonalnym do obecnej populacji (im więcej osobników, tym więcej potomstwa). Zapisz tę zależność w postaci równania różniczkowego.  
Zmiana wartości funkcji względem czasu wyraża się jako pochodna względem czasu t, więc lewa strona równania to:

Obecna populacja to po prostu funkcja P(t), zmiana jest proporcjonalna, a więc potrzebujemy jakiegoś współczynnika proporcjonalności _k_ (z dodatnim znakiem, ponieważ populacja rośnie z czasem). Odpowiedź do zadania:

# Dodatek: słownik

Do sprawnego korzystania z SymPy i wyszukiwania potrzebnych nam funkcji tego pakietu, warto znać anglojęzyczne określenia matematycznych pojęć, które będą pojawiać się na tym przedmiocie:  

- Całka - Integral; całkowanie - integration, to **integrate**  
- Pochodna - Derivative; liczyć pochodną (różniczkować) względem x - to **diff**erentiate with respect to x  
- Równania różniczkowe - Differential Equations  
    - Zwyczajne - Ordinary Differential Equations (ODEs)  
    - Cząstkowe - Partial Differential Equations (PDEs)  
- Rozwiązanie ogólne - general solution; szczególne - particular solution  
- Prawa strona równania - Right-hand side, rhs; Lewa - lhs
- Podstawić wartość (np. podstawić 5 w miejscu zmiennej x) - to **subs**titute 5 for x  
- Stała - constant  
- Funkcja wykładnicza - Exponential function ($f(x) = ab^x$); potęga - exponent.
Uwaga: często pod pojęciem "exponential function", exp() w bibliotekach matematycznych itd. chodzi dokładnie o tzw. funkcję eksponencjalną, czyli $e^x$
